# Countdown Timer

```python
import time
# Step 1: Get user input for countdown start
start = int(input("Enter the number to start the countdown from: "))
# new variable for speed
speed = float(input("Enter the coutdown speed in second (e.g. , 1 for normal, 0.5 for fast): "))
#Calculate the midpoint (We use // to get an integer.)
midpoint = start // 2

# Step 2: Countdown using a whiel loop
print("\n=== Countdown Begins 😊===")
while start > 0:
  print(start)


# Check if it is halfway through.
  if start == midpoint:
    print("We're already halfway there! 😊")

  time.sleep(speed) # < ---Here, we apply the selected speed.
  start -= 1

# Step 3: Print "Happy New Year!"
print("Countdown Complete!😊😊")
```

# Basic Math Quiz Game

```python
import random
import time

# Step 1: Define the math question function
def generate_question():
  num1 = random.randint(1,10)
  num2 = random.randint(1,10)
  operator = random.choice(['+', '-','*','/'])

  if operator == '/':
    # To ensure division is always exact and easy:
    # We generate the result and the divisor first, then calculate the dividend.
    answer = random.randint(1,10) # This will be our clean result.
    num2 = random.randint(1,10) # The number we divide by
    num1 = answer * num2 # The initial large number (e.g., 5 * 3 = 15)
    # So, the question will be "15/3" and the exact answer will be 5.
  else:
    num1 = random.randint(1,10)
    num2 = random.randint(1,10)

    if operator == '+':
      answer = num1 + num2
    elif operator == '-':
      answer = num1 - num2
    else:
      answer = num1 * num2

  return f"{num1} {operator} {num2}", answer

# Step 2: Main Quiz Game Function
def math_quiz():
    #Initial variables
    score = 0
    TIME_LIMIT = 3 # Segundos por pregunta.

    # Welcome message
    print("\n=== Welcome to the Math Quiz Game!😎")
    print("You will presented with math problems, and you need to provide the correct answers.👌")

    # The user decides the number of rounds.
    while True:
        try:
            rounds = int(input("How many rounds do you want to play?: "))
            if rounds > 0:
                break
            print("Please enter a number greater than 0")
        except ValueError:
          print("Invalid input. Please enter an integer.")

    for i in range(rounds):
      question, correct_answer = generate_question()
      print(f"\nQuestion{i + 1}: {question}")
      print(f"¡You have {TIME_LIMIT} seconds to answer")

      start_time = time.time()

      try:
            user_answer = int(input("Your answer: "))
      except ValueError:
            print("Entrada inválida. Se cuenta como incorrecta.")
            user_answer = None

      end_time = time.time()
      enlapsed_time = end_time - start_time

      if enlapsed_time > TIME_LIMIT:
            print(f"⏱️ ¡Tiempo agotado! Tardaste {enlapsed_time:.1f} segundos.")
            print(f"The correct answer was: {correct_answer}")
      elif user_answer == correct_answer:
            print(f"Correct! (Respondido en {enlapsed_time:.1f} segundos)")
            score += 1
      else:
            print(f"Wrong! 😮 The correct answer is: {correct_answer}")


      if user_answer == correct_answer:
        print("Correct!")
        score += 1
      else:
        print(f"Wrong!😥 The correct answer is: {correct_answer}")

    print("\n=== Game Over!🧨")
    print(f"Your final score is: {score}/{rounds}")
    if score == rounds:
      print("Congratulations!🤑 You got all the questions correct.🙌")
    elif score >= rounds // 2:
      print("Good job❗You did well.")
    else:
      print("Keep practicing! You can do better next time 😥")



# Step 3: Run the Game
math_quiz()
```

# Shopping List App

```python
# Step 1: Initialize an empty shopping list
shopping_list = []

# Step 2: Define the main menu
def show_menu():
  print("\n==== Shopping List Menu ====")
  print("1. View the shopping list")
  print("2. Add an item")
  print("3. Remove an item")
  print("4. Edit an item")  # New option
  print("5. Clear List")
  print("6. Exit")          # Changed from 5 to 6

#Step 3: Main Program Loop
while True:
  show_menu()
  choice = input("Enter your choice (1-6): ")

  if choice == "1":
    print("\n=== Shopping List ===")
    if not shopping_list:
      print("Your shopping list is empty.")
    else:
      for index, item in enumerate(shopping_list):
        print(f"{index + 1}. {item}")

  elif choice == "2":
    item = input("Enter the item to add: ")
    shopping_list.append(item)
    print(f"'{item}' has been added to the shopping list.")

  elif choice == "3":
    item = input("Enter the item to remove: ")
    if item in shopping_list:
      shopping_list.remove(item)
      print(f"'{item}' has been removed from the shopping list.")
    else:
      print(f"'{item}' is not in the shopping list.")

  # ---- NEW LOGIC: EDIT AN ARTICLE ----
  elif choice == "4":
    if not shopping_list:
      print("The list is empty. Nothing to edit.")
    else:
      print("\n=== Select the item number to edit ===")
      for index, item in enumerate(shopping_list):
        print(f"{index + 1}. {item}")

      try:
        position = int(input("Enter the number of the item you want to change: ")) - 1
        if 0 <= position < len(shopping_list):
          old_item = shopping_list[position]
          new_item = input(f"Enter the new name for '{old_item}': ")
          shopping_list[position] = new_item
          print(f"'{old_item}' has been changed to '{new_item}'.")
        else:
          print("Invalid item number.")
      except ValueError:
        print("Please enter a valid number.")

  elif choice == "5":
    shopping_list.clear()
    print("The shopping list has been cleared.")

  # ---- MODIFIED: SAVE AND EXIT ----
  elif choice == "6":
    # Save the list to a text file before exiting.
    try:
      with open("lista_compras.txt", "w", encoding="utf-8") as file:
        for item in shopping_list:
          file.write(f"{item}\n")
      print("Shopping list saved successfully to 'lista_compras.txt'.")
    except Exception as e:
      print(f"Error saving the file: {e}")

    print("Goodbye! Happy Shopping!")
    break

  else:
    print("Invalid choice. Please try again.")
```
