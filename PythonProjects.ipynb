### Mini Python projects

```python
import random

secret_number = random.randint(1,10)
attempts = 3

print("I am thinking of a number between 1 and 10")
print("You have three attempts to guess it.😊")

while attempts > 0:
  guess = int(input("Take a Guess: "))
  if guess == secret_number:
    print("Congratulations! You guessed the number:😲")
    break
  elif guess < secret_number:
    print("Too low! Try again.")
  else:
    print("Too high! Try again.")
  attempts -=1

if attempts == 0:
  print("Sorry, you´ve run out of attemps. The secret number was:", secret_number)
```

# Number Comparison Tool


```python
while True: # <-- starts the infinite loop
  # Step 1: Get user input for two numbers
  num1 = float(input("Enter the first number: "))
  num2 = float(input("Enter the second number: "))
  num3 = float(input("Enter the third number: "))

  #Step 2: Compare the numbers and print the result
  print("\n ====== Comparison Results ===== ")
  if num1 == num2 == num3:
    print(f"All three numbers are equal {num1}")
  elif num1 >= num2 and num1>= num3:
    print(f"{num1} is greatest number")
  elif num2 >= num1 and num2>= num3:
    print(f"{num2} is greatest number")
  else:
    print(f"{num3} is greatest number")

  # Step 3: Check if any number is zero
  if num1 == 0 or num2 == 0 or num3==0:
      print("\n least one number is zero.")
  else:
    print("\n Both numbers are non-zero.")

  # --- NEW REPEAT OPTION ---
  print("\n" + "="*30)
  repetir = input("Want to compare other numbers? (y/n): ").strip().lower()

  if repetir != 's': # If the user enters anything other than 's', the loop breaks.
      print("Thank you for using the program! Goodbye. ")
      break  # <--- Break the loop and terminate execution. ---
  print("\n" + "="*30 + "n") # Divider line for the next execution
```
