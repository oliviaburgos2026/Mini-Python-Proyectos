# Countdown Timer

```python
import time
# Step 1: Get user input for countdown start
start = int(input("Enter the number to start the countdown from: "))
# new variable for speed
speed = float(input("Enter the coutdown speed in second (e.g. , 1 for normal, 0.5 for fast): "))
#Calculate the midpoint (We use // to get an integer.)
midpoint = start // 2

# Step 2: Countdown using a whiel loop
print("\n=== Countdown Begins 😊===")
while start > 0:
  print(start)


# Check if it is halfway through.
  if start == midpoint:
    print("We're already halfway there! 😊")

  time.sleep(speed) # < ---Here, we apply the selected speed.
  start -= 1

# Step 3: Print "Happy New Year!"
print("Countdown Complete!😊😊")
```

# Basic Math Quiz Game

```python
import random
import time

# Step 1: Define the math question function
def generate_question():
  num1 = random.randint(1,10)
  num2 = random.randint(1,10)
  operator = random.choice(['+', '-','*','/'])

  if operator == '/':
    # To ensure division is always exact and easy:
    # We generate the result and the divisor first, then calculate the dividend.
    answer = random.randint(1,10) # This will be our clean result.
    num2 = random.randint(1,10) # The number we divide by
    num1 = answer * num2 # The initial large number (e.g., 5 * 3 = 15)
    # So, the question will be "15/3" and the exact answer will be 5.
  else:
    num1 = random.randint(1,10)
    num2 = random.randint(1,10)

    if operator == '+':
      answer = num1 + num2
    elif operator == '-':
      answer = num1 - num2
    else:
      answer = num1 * num2

  return f"{num1} {operator} {num2}", answer

# Step 2: Main Quiz Game Function
def math_quiz():
    #Initial variables
    score = 0
    TIME_LIMIT = 3 # Segundos por pregunta.

    # Welcome message
    print("\n=== Welcome to the Math Quiz Game!😎")
    print("You will presented with math problems, and you need to provide the correct answers.👌")

    # The user decides the number of rounds.
    while True:
        try:
            rounds = int(input("How many rounds do you want to play?: "))
            if rounds > 0:
                break
            print("Please enter a number greater than 0")
        except ValueError:
          print("Invalid input. Please enter an integer.")

    for i in range(rounds):
      question, correct_answer = generate_question()
      print(f"\nQuestion{i + 1}: {question}")
      print(f"¡You have {TIME_LIMIT} seconds to answer")

      start_time = time.time()

      try:
            user_answer = int(input("Your answer: "))
      except ValueError:
            print("Entrada inválida. Se cuenta como incorrecta.")
            user_answer = None

      end_time = time.time()
      enlapsed_time = end_time - start_time

      if enlapsed_time > TIME_LIMIT:
            print(f"⏱️ ¡Tiempo agotado! Tardaste {enlapsed_time:.1f} segundos.")
            print(f"The correct answer was: {correct_answer}")
      elif user_answer == correct_answer:
            print(f"Correct! (Respondido en {enlapsed_time:.1f} segundos)")
            score += 1
      else:
            print(f"Wrong! 😮 The correct answer is: {correct_answer}")


      if user_answer == correct_answer:
        print("Correct!")
        score += 1
      else:
        print(f"Wrong!😥 The correct answer is: {correct_answer}")

    print("\n=== Game Over!🧨")
    print(f"Your final score is: {score}/{rounds}")
    if score == rounds:
      print("Congratulations!🤑 You got all the questions correct.🙌")
    elif score >= rounds // 2:
      print("Good job❗You did well.")
    else:
      print("Keep practicing! You can do better next time 😥")



# Step 3: Run the Game
math_quiz()
```

# Shopping List App

```python
# Step 1: Initialize an empty shopping list
shopping_list = []

# Step 2: Define the main menu
def show_menu():
  print("\n==== Shopping List Menu ====")
  print("1. View the shopping list")
  print("2. Add an item")
  print("3. Remove an item")
  print("4. Edit an item")  # New option
  print("5. Clear List")
  print("6. Exit")          # Changed from 5 to 6

#Step 3: Main Program Loop
while True:
  show_menu()
  choice = input("Enter your choice (1-6): ")

  if choice == "1":
    print("\n=== Shopping List ===")
    if not shopping_list:
      print("Your shopping list is empty.")
    else:
      for index, item in enumerate(shopping_list):
        print(f"{index + 1}. {item}")

  elif choice == "2":
    item = input("Enter the item to add: ")
    shopping_list.append(item)
    print(f"'{item}' has been added to the shopping list.")

  elif choice == "3":
    item = input("Enter the item to remove: ")
    if item in shopping_list:
      shopping_list.remove(item)
      print(f"'{item}' has been removed from the shopping list.")
    else:
      print(f"'{item}' is not in the shopping list.")

  # ---- NEW LOGIC: EDIT AN ARTICLE ----
  elif choice == "4":
    if not shopping_list:
      print("The list is empty. Nothing to edit.")
    else:
      print("\n=== Select the item number to edit ===")
      for index, item in enumerate(shopping_list):
        print(f"{index + 1}. {item}")

      try:
        position = int(input("Enter the number of the item you want to change: ")) - 1
        if 0 <= position < len(shopping_list):
          old_item = shopping_list[position]
          new_item = input(f"Enter the new name for '{old_item}': ")
          shopping_list[position] = new_item
          print(f"'{old_item}' has been changed to '{new_item}'.")
        else:
          print("Invalid item number.")
      except ValueError:
        print("Please enter a valid number.")

  elif choice == "5":
    shopping_list.clear()
    print("The shopping list has been cleared.")

  # ---- MODIFIED: SAVE AND EXIT ----
  elif choice == "6":
    # Save the list to a text file before exiting.
    try:
      with open("lista_compras.txt", "w", encoding="utf-8") as file:
        for item in shopping_list:
          file.write(f"{item}\n")
      print("Shopping list saved successfully to 'lista_compras.txt'.")
    except Exception as e:
      print(f"Error saving the file: {e}")

    print("Goodbye! Happy Shopping!")
    break

  else:
    print("Invalid choice. Please try again.")
```

# Contact Book


```python
import json
import os

# Step 1: Initialize the contact book and define the save file
FILE_NAME = "contacts_book.json"
contacts = {}


# Persistence Functions (Save and Load)
def save_contacts():
    """Saves the contacts dictionary to a JSON file."""
    try:
        with open(FILE_NAME, "w", encoding="utf-8") as file:
            json.dump(contacts, file, ensure_ascii=False, indent=4)
        print("💾 Contact book saved successfully!")
    except Exception as e:
        print(f"Error saving contacts: {e}")


def load_contacts():
    """Loads contacts from the JSON file if it exists."""
    global contacts
    if os.path.exists(FILE_NAME):
        try:
            with open(FILE_NAME, "r", encoding="utf-8") as file:
                contacts = json.load(file)
            print("📂 Contact book loaded successfully!")
        except Exception as e:
            print(f"Error loading contacts: {e}. Starting with an empty book.")
    else:
        contacts = {}


# Step 2: Display the menu
def show_menu():
    print("\n===== Contact Book Menu =====")
    print("1. Add Contact")
    print("2. View Contacts")
    print("3. Search Contact")
    print("4. Edit Contact")
    print("5. Delete Contact")
    print("6. Exit")


# Step 3: Add a Contact (Supports multiple values separated by commas)
def add_contact():
    name = input("Enter contact name: ")

    # Ask for inputs and convert them into lists using split(',')
    raw_phones = input("Enter phone numbers (separated by commas): ")
    raw_emails = input("Enter emails (separated by commas): ")

    # .strip() removes unnecessary whitespace around each element
    phones = [p.strip() for p in raw_phones.split(",") if p.strip()]
    emails = [e.strip() for e in raw_emails.split(",") if e.strip()]

    contacts[name] = {"phones": phones, "emails": emails}
    print(f"Contact {name} has been added successfully! ❗👌")
    save_contacts()  # Automatically save after adding


# Step 4: View All Contacts
def view_contacts():
    if contacts:
        print("\n===== Contacts =====")
        for name, details in contacts.items():
            # Use ', '.join() to format the list cleanly for display
            print(f"\nName:   {name}")
            print(f"Phones: {', '.join(details['phones'])}")
            print(f"Emails: {', '.join(details['emails'])}")
            print("-" * 20)
    else:
        print("Your contact book is empty 😰")


# Step 5: Search a Contact
def search_contact():
    name = input("Enter the name of the contact you want to search: ")
    if name in contacts:
        print(f"\n===== Contact Details for {name} =====")
        print(f"Name:   {name}")
        print(f"Phones: {', '.join(contacts[name]['phones'])}")
        print(f"Emails: {', '.join(contacts[name]['emails'])}")
    else:
        print(f"Contact {name} not found in your contact book. 😨")


# Step 6: Edit a contact
def edit_contact():
    name = input("Enter the name of the contact you want to edit: ")
    if name in contacts:
        print(f"\n===== Editing Contact: {name} =====")
        raw_phones = input(
            "Enter new phone numbers (separated by commas) or press Enter to keep current: "
        )
        raw_emails = input(
            "Enter new emails (separated by commas) or press Enter to keep current: "
        )

        # If the user input is empty, retain the existing values
        if raw_phones.strip():
            contacts[name]["phones"] = [
                p.strip() for p in raw_phones.split(",") if p.strip()
            ]
        if raw_emails.strip():
            contacts[name]["emails"] = [
                e.strip() for e in raw_emails.split(",") if e.strip()
            ]

        print(f"Contact {name} has been updated successfully. 👍")
        save_contacts()  # Automatically save after editing
    else:
        print(f"Contact {name} not found in your contact book. 😨")


# Step 7: Delete a contact
def delete_contact():
    name = input("Enter the name of contact you want to delete: ")
    if name in contacts:
        del contacts[name]
        print(f"Contact {name} has been deleted successfully! 👍")
        save_contacts()  # Automatically save after deleting
    else:
        print(f"Contact {name} not found in your contact book. 😨")


# Step 8: Main Program Loop
# Load existing contacts at application startup
load_contacts()

while True:
    show_menu()
    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        add_contact()
    elif choice == "2":
        view_contacts()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        edit_contact()
    elif choice == "5":
        delete_contact()
    elif choice == "6":
        print("Thank you for using the Contact Book. Goodbye! 😎")
        break
    else:
        print("Invalid choice. Please select a valid option (1-6). 😨")
```

# Ingredients Checker

```python
import json
import os

# File name where recipes will be stored
RECIPES_FILE = "recipes.json"

# Default recipe database (used if the file doesn't exist yet)
DEFAULT_RECIPES = {
    "Basic Cake": ["flour", "sugar", "butter", "eggs", "milk"],
    "Cookies": ["flour", "sugar", "butter"],
    "Flan": ["milk", "eggs", "sugar"]
}

# Dictionary for alternative ingredients (Substitutes)
# Key: required ingredient -> Value: accepted alternative
SUBSTITUTES = {
    "milk": "almond milk",
    "sugar": "stevia",
    "butter": "margarine"
}

# =====================================================================
# FUNCTIONS: SAVE AND LOAD FROM FILE
# =====================================================================
def load_recipes():
    """Loads recipes from the JSON file or initializes it with defaults."""
    if os.path.exists(RECIPES_FILE):
        with open(RECIPES_FILE, "r", encoding="utf-8") as f:
            # Load file and convert inner lists back into sets
            data = json.load(f)
            return {name: set(ingredients) for name, ingredients in data.items()}
    else:
        # If the file doesn't exist, save defaults first
        save_recipes(DEFAULT_RECIPES)
        return {name: set(ingredients) for name, ingredients in DEFAULT_RECIPES.items()}

def save_recipes(recipes_dict):
    """Saves the recipes dictionary into the JSON file."""
    # JSON cannot save 'sets' directly, so we convert them to lists
    data_to_save = {name: list(ingredients) for name, ingredients in recipes_dict.items()}
    with open(RECIPES_FILE, "w", encoding="utf-8") as f:
        json.dump(data_to_save, f, indent=4, ensure_ascii=False)


# =====================================================================
# MAIN PROCESS
# =====================================================================

# Load recipes at startup
recipes = load_recipes()

# Get available ingredients from the user (trimming spaces and converting to lowercase)
user_input = input("Enter the ingredients you have (separated by commas): ")
user_ingredients = set(ing.strip().lower() for ing in user_input.split(",") if ing.strip())

print("\n===== Ingredients Check Results =====")
possible_recipes = 0

# Iterate through every saved recipe
for recipe_name, recipe_ingredients in recipes.items():

    # Core comparison (set difference)
    missing_ingredients = recipe_ingredients - user_ingredients

    # Evaluate alternative ingredients for what is missing
    missing_real = set()
    used_alternatives = []

    for missing in missing_ingredients:
        # If the missing ingredient has a substitute and the user HAS that substitute...
        if missing in SUBSTITUTES and SUBSTITUTES[missing] in user_ingredients:
            used_alternatives.append(f"{missing} ➡️ use {SUBSTITUTES[missing]}")
        else:
            # If no substitute is available, it is genuinely missing
            missing_real.add(missing)

    # Suggest recipe based on the final check
    if not missing_real:
        possible_recipes += 1
        print(f"\n✨ You can make: **{recipe_name}** 👌")
        if used_alternatives:
            print(f"   💡 Alternative note: {', '.join(used_alternatives)}")

    elif len(missing_real) == 1:
        # Optional: Let the user know if they are just one ingredient away
        print(f"\n🔸 For **{recipe_name}** you are missing only: {', '.join(missing_real)} 😨")

if possible_recipes == 0:
    print("\nYou don't have enough ingredients for any complete recipe right now. 😢")

# --- Optional Block: Add a new recipe to test data persistence ---
print("\n" + "-"*40)
save_new = input("Want to add a new recipe? (yes/no): ").strip().lower()
if save_new in ["yes", "y"]:
    new_name = input("Recipe name: ").strip()
    new_ings = input("Ingredients (separated by commas): ")
    set_ing = set(i.strip().lower() for i in new_ings.split(",") if i.strip())

    # Add to current dictionary and save to disk
    recipes[new_name] = set_ing
    save_recipes(recipes)
    print(f"✅ '{new_name}' has been saved to {RECIPES_FILE}!")
```
